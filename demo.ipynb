```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import json
import random
import time
import fake_useragent
```

```python
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
}
proxies={
    "http":"http://175.42.158.121:9999"
}
ses = requests.session()
response = ses.get('https://api.bilibili.com/x/space/acc/info?mid=546195&jsonp=jsonp', headers=headers)
```

```python
def get_response(url):
    response = ses.get(url)
    if response.status_code == 412:
        print('被禁用IP了')
    elif response.status_code != 200:
        print('出错了！', response.status_code)
    else:
        return json.loads(response.text)
    # exit()
    return {}
```

```python
def get_up_info(mid):
    up_info = {}
    base_info_url = 'https://api.bilibili.com/x/space/acc/info?mid={}&jsonp=jsonp'.format(mid)
    dic = get_response(base_info_url)
    up_info['mid'] = dic['data']['mid']
    up_info['name'] = dic['data']['name']
    up_info['sex'] = dic['data']['sex']
    up_info['level'] = dic['data']['level']
    up_info['has_fans_badge'] = dic['data']['fans_badge']
    up_info['title'] = dic['data']['official']['title']
    up_info['has_live_room'] = 1 if dic['data']['live_room']['roomStatus'] else 0

    follow_url = "https://api.bilibili.com/x/relation/stat?vmid={}&jsonp=jsonp".format(mid)
    dic = get_response(follow_url)
    if dic:
        up_info['following'] = dic['data']['following']
        up_info['follower'] = dic['data']['follower']
    
    view_like_url = "https://api.bilibili.com/x/space/upstat?mid={}&jsonp=jsonp".format(mid)
    dic = get_response(view_like_url)
    if dic:
        up_info['view'] = dic['data']['archive']['view']
        up_info['likes'] = dic['data']['likes']

    video_url = "https://api.bilibili.com/x/space/arc/search?mid={}&ps=30&tid=0&pn=1&keyword=&order=pubdate&jsonp=jsonp".format(mid)
    dic = get_response(video_url)
    tlist = {}
    zone = dic['data']['list']['tlist']
    if zone:
        # TODO: 这里用分区的name做key，还是分区的id做key？
        for key in zone.keys():
            # tlist[key] = zone[key]['count'] # id
            tlist[zone[key]['name']] = zone[key]['count'] # name
    up_info['zone'] = tlist
    # print(up_info)
    time.sleep(1+random.random())
    return up_info
```

```python
up_list = pd.read_csv('up_list_raw.csv')
up_mid_list = up_list.mid
mid_map = {}
for idx, mid in enumerate(up_mid_list):
    if idx == 3000:  break
    print('*'*10,'index = ',idx, '*'*10)
    mid_map[mid] = get_up_info(mid)

```

```python
df = pd.DataFrame(mid_map).T
df.to_csv('up_info.csv')
```
